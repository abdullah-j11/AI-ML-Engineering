```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = sns.load_dataset('iris')
```

```python
print('shape (rows, columns):', df.shape)
```

```python
print('Column Names:', df.columns.tolist())
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    data=df,
    x='sepal_length',
    y='petal_length',
    hue='species',
    palette='Set1',
    s=80
)
plt.title('Scatter Plot: Sepal length vs Petal length', fontsize=14)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend(title='Species')
plt.tight_layout()
plt.show()
```

```python
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

fig, axes = plt.subplots(2, 2, figsize=(10, 7))

for ax, feature in zip(axes.flatten(), features):
    sns.histplot(data=df, x=feature, hue='species', kde=True,
                 palette='Set1', ax=ax, bins=15)
    ax.set_title(f'Distribution of {feature.replace("_", " ").title()}')
    ax.set_xlabel(feature.replace('_', ' ').title() + ' (cm)')

plt.suptitle('Feature Distributions by Species', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 7))
for ax, feature in zip(axes.flatten(), features):
    sns.boxplot(data=df, x='species', y=feature,
                palette='Set1', ax=ax)
    ax.set_title(f'Box Plot: {feature.replace("_", " ").title()}')
    ax.set_xlabel('Species')
    ax.set_ylabel(feature.replace('_', ' ').title() + ' (cm)')

plt.suptitle('Box Plots — Outlier Detection by Species', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```
